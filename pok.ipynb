```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(image_data,l, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

lr.fit(X_train,y_train)
ans=lr.predict(X_test)
np.mean(ans==y_test)
```

```python
lr_ans=lr.predict(image_data_test)
lr_ans.shape

lr_ans=le.inverse_transform(lr_ans)
df10=pd.DataFrame({'ImageId':label_name_test,'NameOfPokemon':lr_ans})
df7=pd.merge(df9,df10,on='ImageId')
df10.to_csv('poke_lr.csv',index=False)
df=pd.read_csv('poke_lr.csv')
df
```
