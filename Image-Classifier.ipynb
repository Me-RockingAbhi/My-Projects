```
import numpy as np
p=Path("/content/gdrive/My Drive/Testimg")
dir=p.glob("*")
image_data1=[]
img_label1=[]
c=0;
for i in dir:
    name=(str(i).split("\\")[-1])
    pic=image.load_img(i,target_size=(299,299))
    i_ar=image.img_to_array(pic)
    image_data1.append(i_ar)
    print(name,c)
    img_label1.append(name)
    c=c+1;
xte=np.array(image_data1)/255.0
```

```
print(img_label1[0])
```

```
from keras.applications import DenseNet169
from keras import models
from keras.models import Sequential
from keras.layers import Dropout,Dense,Convolution2D, GlobalAveragePooling2D,Activation
model=Sequential([DenseNet169(include_top=False,input_shape=(299,299,3)),
                  Dropout(0.3),
                  Convolution2D(4, (3, 3), padding='valid'),
                  Activation('relu'),
                  GlobalAveragePooling2D(),
                  Activation('softmax')])
```

```
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
le=LabelEncoder()
data2['Class']=le.fit_transform(data2['Class'])
y_tr=data2['Class']
print(y_tr.shape)
y=np.array(y_tr)
y=np_utils.to_categorical(y)
```

```
from keras.optimizers import Adam
model.compile(
    optimizer=Adam(lr=0.0001),
    loss='mean_squared_error',
    metrics=['accuracy']
)
model.fit(x,y,
          epochs=12,
          verbose=1,)
```

```
tdata=pd.read_csv("/content/gdrive/My Drive/test.csv")
```

```

ans=model.predict_classes(xte)
ans1=list(le.inverse_transform(ans))
print(len(ans1))
print(len(img_label1))
solution=pd.DataFrame()
la=[]
for i in img_label1:
  name=(str(i).split("/")[-1])
  la.append(name)
yr=np.array(la)
solution['Image']=yr
solution['Class']=ans1
print(solution.head)
print(tdata.head)
solu=pd.merge(tdata,solution,on='Image')
solu.to_csv('Submit2.csv',header=True,index = None)
```
